```python
import numpy as np
import matplotlib.pyplot as plt

def generar_datos(N):
    sexo = np.random.choice(['MASCULINO', 'FEMENINO', 'NO ESPECIFICA'], N ,[0.4, 0.5, 0.1])  # Por ejemplo, 40% masculino, 50% femenino, 10% no especificado)
    # Generar edades con distribución normal
    edad =np.random.normal(44, 6, N)
    edad=np.clip(np.around(edad),28,60)
    # Generar alturas, pesos y otras características
    altura = np.random.normal(180, 10, N)
    altura = np.clip(np.around(altura, decimals=1), 150.0, 210.0)

    peso = np.random.normal(80, 15, N)
    peso = np.clip(np.around(peso, decimals=1), 40.0, 130.0)

    horas_trabajo_estudio = np.random.normal(10, 2, N)
    horas_trabajo_estudio = np.clip(np.around(horas_trabajo_estudio, decimals=1), 4.0, 12.0)

    horas_ejercicio = np.random.normal(1, 0.5, N)
    horas_ejercicio = np.clip(np.around(horas_ejercicio, decimals=1), 0, 2)

    horas_descanso = 24 - horas_trabajo_estudio - horas_ejercicio
    datos = np.column_stack((sexo, edad, altura, peso, horas_descanso, horas_trabajo_estudio, horas_ejercicio))
    return datos


# Solicitar al usuario la cantidad N de datos a cargar
N = 1000

# Generar datos aleatorios para N personas con distribución normal y límites específicos
datos = generar_datos(N)
```

```python
import pandas as pd
# Crear un DataFrame con los datos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear un DataFrame con los datos de ventas
data = {
    'Dia': ['Lunes', 'Martes', 'Miercoles', 'Jueves', 'Viernes', 'Sabado', 'Domingo'],
    'Producto_A': [150, 130, 160, 140, 170, 180, 155],
    'Producto_B': [80, 90, 85, 70, 95, 100, 110],
    'Producto_C': [200, 220, 210, 190, 230, 240, 210],
    'Producto_D': [120, 150, 130, 140, 160, 140, 130],
    'SubTotal': [750, 590, 585, 540, 655, 660, 615]
}

df = pd.DataFrame(data)

# Calcular la media y mediana de ventas por producto
media_por_producto = df.drop('Dia', axis=1).iloc[:, :-1].mean()
mediana_por_producto = df.drop('Dia', axis=1).iloc[:, :-1].median()

# Plotear gráfico de barra para la media y mediana por producto
plt.figure(figsize=(10, 6))
media_por_producto.plot(kind='bar', color='blue', label='Media')
mediana_por_producto.plot(kind='bar', color='orange', label='Mediana')
plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.title(' Ventas por Producto')
plt.legend()
plt.show()

# Calcular la media y mediana de ventas por día
media_por_dia = df.drop(columns=['Dia', 'SubTotal']).mean(axis=1)
mediana_por_dia = df.drop(columns=['Dia', 'SubTotal']).median(axis=1)

plt.figure(figsize=(10, 6))
plt.plot(df['Dia'], media_por_dia, marker='o', color='blue', label='Media')
plt.plot(df['Dia'], mediana_por_dia, marker='o', color='red', label='Mediana')
plt.xlabel('Día')
plt.ylabel('Ventas')
plt.title(' Ventas por Día')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()


desvio_por_producto = df.drop(['Dia', 'SubTotal'], axis=1).std()


producto_mayor_desvio = desvio_por_producto.idxmax()
producto_menor_desvio = desvio_por_producto.idxmin()

# Plotear las ventas de los productos con mayor y menor desvío estándar
plt.figure(figsize=(10, 6))
df.plot(x='Dia', y=[producto_mayor_desvio, producto_menor_desvio], kind='bar')
plt.xlabel('Día')
plt.ylabel('Ventas')
plt.title(f' Productos con Mayor y Menor Desvío Estándar')
plt.xticks(rotation=45)
plt.legend([f'{producto_mayor_desvio} (Mayor Desvío)', f'{producto_menor_desvio} (Menor Desvío)'])
plt.show()


```
