Simular el lanzamiento de un dado eqilibrado de 6 caras.

*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5

```python
import numpy as np
import matplotlib.pyplot as plt

probabilidades = [0.4, 0.4, 0.1, 0.1]
probabilidades = [p / sum(probabilidades) for p in probabilidades]


num_lanzamientos = int(input("Ingrese el número de lanzamientos: "))

resultados = np.random.choice([1, 2, 3, 4], num_lanzamientos, p=probabilidades)

prob_2 = np.sum(resultados == 2) / num_lanzamientos
prob_3 = np.sum(resultados == 3) / num_lanzamientos
prob_5 = np.sum(resultados == 5) / num_lanzamientos

print(f"Probabilidad de obtener el número 2: {prob_2:.4f}")
print(f"Probabilidad de obtener el número 3: {prob_3:.4f}")
print(f"Probabilidad de obtener el número 5: {prob_5:.4f}")

valores = [2, 3, 5]
probabilidades_resultados = [prob_2, prob_3, prob_5]

plt.bar(valores, probabilidades_resultados)
plt.xlabel('Número')
plt.ylabel('Probabilidad')
plt.title('Histograma de Probabilidades de Lanzamiento de Dado No Equilibrado')
plt.xticks(valores)
plt.show()


```

 Realizar la simulación del juego {piedra,papel , tijera} donde el resultado de esta simulación deberá ser lo que mi oponente elije. No debe considerarse la simulación de mi elección.



*   Calcular la probailidad aproximada que mi oponente elija, 'piedra', 'papel', o 'tijera', en N jugadas.
*   Realizar un grafico de barra mostrando estas probabilidad. Concluir algo sobre las probabilidades de estos resultados, cuando el número de jugadas se incrementa bastante.




```python
import numpy as np
import matplotlib.pyplot as plt


opciones = ['piedra', 'papel', 'tijera']

num_jugadas = int(input("Ingrese el número de jugadas: "))

elecciones_oponente = np.random.choice(opciones, num_jugadas)

prob_piedra = np.sum(elecciones_oponente == 'piedra') / num_jugadas
prob_papel = np.sum(elecciones_oponente == 'papel') / num_jugadas
prob_tijera = np.sum(elecciones_oponente == 'tijera') / num_jugadas

print(f"Probabilidad de que el oponente elija 'piedra': {prob_piedra:.4f}")
print(f"Probabilidad de que el oponente elija 'papel': {prob_papel:.4f}")
print(f"Probabilidad de que el oponente elija 'tijera': {prob_tijera:.4f}")


valores = ['piedra', 'papel', 'tijera']
probabilidades_resultados = [prob_piedra, prob_papel, prob_tijera]

plt.bar(valores, probabilidades_resultados)
plt.xlabel('Elección del Oponente')
plt.ylabel('Probabilidad')
plt.title('Probabilidades de Elecciones del Oponente en el Juego de Piedra, Papel o Tijera')
plt.show()
```
