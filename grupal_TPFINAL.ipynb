```python
#valores atipicos
url = '/content/world-data-2023.csv'
data = pd.read_csv(url)

# Calcular el rango intercuartílico (IQR)
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Definir límites para identificar valores atípicos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Encontrar valores atípicos (outliers)
outliers = (data < lower_bound) | (data > upper_bound)

# Mostrar los valores atípicos
print("Valores atípicos (outliers):")
print(outliers)

# Contar la cantidad de valores atípicos por columna
outliers_count = outliers.sum()
print("Cantidad de valores atípicos por columna:")
print(outliers_count)

```

```python
# 5 - Se realiza un Análisis Univariado en base a los siguientes datos de la tabla:
# Contry y Co2 emission, urban population y population Labor force participation
# Realizar análisis univariado (Histogramas)
url = '/content/world-data-2023.csv'
data = pd.read_csv(url)

# Definimos las columnas donde tomamos los datos
columnas_analisis = ['Country', 'Co2-Emissions', 'Urban_population']

# Análisis univariado
for columna in columnas_analisis:
    # Gráfico para mostrar la frecuencia de cada valor único
    plt.figure(figsize=(8, 5))
    data[columna].value_counts().plot(kind='bar')
    plt.title(f'Gráfico de frecuencia de {columna}')
    plt.xlabel(columna)
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Gráfico de torta
    plt.figure(figsize=(8, 5))
    counts = data[columna].value_counts()
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    plt.axis('equal')
    plt.title(f'Gráfico de torta de {columna}')
    plt.tight_layout()
    plt.show()
```

EJERCICO 6-
Entre las variables seleccionadas obtenemos valor de -0.29, es una correlación negativa moderada y sugiere que a medida que aumenta el valor de variable "Density (P/Km2), hay una tendencia a que el valor de la otra variable "Co2-Emissions" disminuya, y viceversa.
Esto podría estar relacionado con el hecho de que áreas con alta densidad de población podrían tener un mayor enfoque en la eficiencia energética y en la implementación de tecnologías más limpias para reducir las emisiones.





```python
# ejercicio 6
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = '/content/world-data-2023.csv'
data = pd.read_csv(url)

# Definir las dos columnas para la correlación
columna_enteros = 'Density\n(P/Km2)'
columna_decimales = 'Co2-Emissions'
data[columna_enteros] = pd.to_numeric(data[columna_enteros], errors='coerce')
data[columna_decimales] = pd.to_numeric(data[columna_decimales], errors='coerce')
data = data.dropna(subset=[columna_enteros, columna_decimales])

# Calcular la correlación entre las dos columnas
correlation = data[columna_enteros].corr(data[columna_decimales])

#correlación
print(f'Correlación entre {columna_enteros} y {columna_decimales}: {correlation:.2f}')
plt.figure(figsize=(8, 5))
plt.scatter(data[columna_enteros], data[columna_decimales])
plt.title(f'Correlación entre {columna_enteros} y {columna_decimales}: {correlation:.2f}')
plt.xlabel(columna_enteros)
plt.ylabel(columna_decimales)
plt.grid(True)
plt.show()
# Calcular matriz de correlación
matriz_correlacion = data[[columna_enteros, columna_decimales]].corr()

# Crear mapa de calor de la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlación entre Density y Co2-Emissions')
plt.show()

```
